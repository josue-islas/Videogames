```python
import pandas as pd
import yfinance as yf
```

```python
def extracción(x, inicio='2020-01-01', termino='2022-01-01'):
    df = yf.download(x, start=inicio, end=termino, progress=False)
    close_TSLA = df['Close']
    diccionario = {'Close':close_TSLA.values}
    df_close = pd.DataFrame(diccionario)
    return df_close
```

```python
empresas = ['A',
'AAL',
'AAP',
'AAPL',
'ABBV',
'ABC',
'ABMD',
'ABT',
'ACN',
'ADBE',
'ADI',
'ADM',
'ADP']

len(empresas)
```

```python
df_sp = pd.DataFrame()
for i in empresas:
    df_sp[i] = extracción(i)

df_sp.head()
```

```python
df_sp.plot(subplots = True, figsize=(12,17))
```

```python
df_sp.mean()
```

```python
df_sp.std()
```

```python
df = ts.ExtraerDatos(['AMZN'])
serie = df.iloc[:,0]
df.head()
```

```python
ts.fuller_test(serie)
```

```python
ts.autocorrelation_plots(serie)
```

```python
from statsmodels.tsa.arima.model import ARIMA

model = ARIMA(serie, order=(3,1,5))
model_fit = model.fit()
print(model_fit.summary())
```

```python
ts.residual_plot(model_fit.resid)
```

```python
import matplotlib.pyplot as plt
model_fit.plot_diagnostics(figsize=(15,9))
plt.plot()
```
